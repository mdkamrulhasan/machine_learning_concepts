Define data class

```python
class MyNode:
  def __init__(self, x):
    self.data = x
    self.next_node = None
```

Define your Linked List

```python
class MyLinkedList:
  def __init__(self):
    self.start = None

  # adding a data item
  def insert(self, x):
    node_x = MyNode(x)
    #node_x.data = x
    node_x.next_node = self.start
    self.start = node_x
    #

  # Traversing over the linked list
  def traverse(self):
    ptr = self.start
    while ptr != None:
      print(ptr.data)
      ptr = ptr.next_node

  # Remove from list
  def delete(self):
    self.start = self.start.next_node
```

Instantiate your Linked List

```python
my_ll = MyLinkedList()
```
